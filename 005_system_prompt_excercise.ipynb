```python
# Create an API client

from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-6"
```

```python
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

def chat(messages, system=None):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
    }
    
    if system:
        params["system"] = system
    
    message = client.messages.create(**params)
    return message.content[0].text
```

```python
messages = []

system_prompt = """
You are a Python engineer who writes very consise code
"""

add_user_message(messages, "Write a Python function that checks a string for duplicate characters.")

answer_without_system_prompt = chat(messages)

answer_with_system_prompt = chat(messages, system=system_prompt)
```

```python
# Output the answer without system prompt

print (answer_without_system_prompt)
```

```python
# Output the answer with system prompt

print (answer_with_system_prompt)
```
